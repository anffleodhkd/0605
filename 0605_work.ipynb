```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from scipy.stats import zscore
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
c1 = pd.read_csv('C:/DIPython/concrete.csv')
```

```python
c1
```

```python
def zscore_standize(c1):
    mean = c1.mean()
    std = c1.std()
    return (c1 - mean) / std
```

```python
c1.apply(zscore_standize)
```

```python
c1.apply(zscore_standize).plot()
```

```python
def minmax_normalize(c1):
    xmax = c1.max()
    xmin = c1.min()    
    return (c1 - xmin) / (xmax-xmin)
```

```python
c1.apply(minmax_normalize)
```

```python
c1.apply(minmax_normalize).plot()
```
